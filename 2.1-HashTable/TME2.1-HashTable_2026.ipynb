1\. Écrivez une fonction pour générer aléatoirement des séquences artificielles. Utilisez les fonctions de la bibliothèque `random`. Toutes les nuclétodies sont équiprobables pour la génération des séquences. Si le nombre de séquences est égal à 1 (`n = 1`), renvoyez une chaîne (string) de caractères au lieu d'une liste.

```python
import random

nuc = ('A', 'C', 'G', 'T')

# Pour faire un test simple
n = 3 # nombre de séquences
t = 8 # taille des séquences

def generateRandomSequences(n:int, t:int, upper = True):
    """
    Génére plusieurs séquences nucléotidiques aléatoires 
    entrée n : nombre de séquences
    entrée t : taille des séquences
    entrée upper : bool, si True, les nucléotides seront majuscule, False, minuscule
    sortie séquences : liste de séquences nucléotidiques aléatoires ou un string si n = 1
    """
    sequences = [] # liste des séquences nucléotidiques 
    for i in range(n) :
        sequence = "" 
        for j in range(t) : 
            # ajoute t nucléotides aléatoirement
            sequence += nuc[random.randint(0, 3)]
        sequences.append(sequence)
    return sequences

generateRandomSequences(n = n, t = t, upper = True)
```

```python
k = 4 # taille des motifs

# Pour le bonus
f = 0.9 # fréquence d'occurences des motifs

def implantMotifs(motif:str, sequences:list, f = 0.9):
    """
    Insére un motif à des positions aléatoires des séquences
    entrée motif : motif qui va être implanté dans les séquences
    entrée séquences : liste de séquences
    entrée f : fraction des séquences qui contiendront le motif
    sortie modified_sequences: liste de séquences ayant le motif implanté
    """
    n = len(sequences) # nombre de séquences 
    nb_implant = int(f * n) # nombre de séquences comportant le motif 
    # Copie des séquences pour les modifier
    modified_sequences = sequences.copy()
    # choix UNIQUE des indices à modifier (il n'y a pas de doublon avec cette fonction)
    indices_modifies = random.sample(population = range(n), k = nb_implant) # pour n séquences en choisir aléatoirement f*n
    
    for i in indices_modifies :
        sequence = modified_sequences[i] # séquence à modifier 
        pos = random.randint(0, len(sequence)) # position uniforme
        modified_sequences[i] = sequence[:pos] + motif + sequence[pos:] # insertion 

    return modified_sequences

# Pour obtenir le même résultat que dans le dévelopent de ce TME
random.seed(seed)

motif_artificiel: str =  generateRandomSequences(1, k)[0]
seq_artificielles =  generateRandomSequences(10, 41)

print("Le motif géneré est : ", motif_artificiel)

print("\nSéquences artificielles :")
for seq in seq_artificielles :
    print(seq)
    
print("\nSéquences modifiées :")
for seq in implantMotifs(motif_artificiel, seq_artificielles, f) : 
    print(seq)
```

4\. Faites une fonction pour chercher les $m$ motifs de taille $k$ les plus fréquents dans l'ensemble des séquences.

```python
def searchMotifs(k:int, sequences:list):
    """
    Cherche les motifs de taille k dans un ensemble de séquences
    entrée k : taille du motif
    entrée séquences : liste de séquences
    sortie motifs: dictionnaire de motifs, clé = motif, valeur = fréquence d'observation
    >>>searchMotifs(3, ['TAAGTAA', 'TATAA', 'CTATC'])
    {'TAA': 3, 'AAG': 1, 'AGT': 1, 'GTA': 1, 'TAT': 2, 'ATA': 1, 'CTA': 1, 'ATC': 1}
    """
    motifs  = {} # dictionnaire des motifs (motif : occurence)
    for seq in sequences :
        for i in range(len(seq) - k + 1) : 
            motif = seq[i : i + k]
            if motif in motifs :
                motifs[motif] += 1
            else:
                motifs[motif] = 1
    return motifs

test_motifs = searchMotifs(3, ['TAAGTAA', 'TATAA', 'CTATC'])

print(test_motifs)
```

5\. Faite aussi une fonction qui affiche les `top` motifs les plus fréquents. Cette fonction peut également trier le dictionnaire de manière croissant ou décroissant.  
*Note*: Voir les paramètres `key` et `reverse` de la fonction `sorted`

```python
def getTopMotifs(motifs:dict, top, decroissant = True):
    """
    Renvoie les top motifs le plus fréquent
    entrée motifs: dictionnaire de motifs, clé = motif, valeur = fréquence d'observation
    entrée top : les top plus fréquent, si top == "all", renvoie le dictionnaire complet trié sans filtrer
    entrée decroissant : bool, si True le dictionnaire est trié par ordre décroissant de valeur
    sortie motifsfreq: dictionnaire contenant les top motifs les plus fréquents, clé = motif, valeur = fréquence d'observation
    
    >>>getTopMotifs({'TAA': 3, 'AAG': 1, 'AGT': 1, 'GTA': 1, 'TAT': 2, 'ATA': 1, 'CTA': 1, 'ATC': 1}, 2)
    {'TAA': 3, 'TAT': 2}
    """
    # Tri du dictionnaire par fréquence en retournant une liste :
        # liste tuples (cles,val)
        # sur quoi trier : ici les valeurs donc le deuxième élèment du tuple (fonction anonyme)
        # en ordre décroissant (ou pas)
    sorted_motifs = sorted(motifs.items(), key = lambda x: x[1], reverse =  decroissant)

    if top != "all" : 
        # Filtrer selon le nombre top demandé (déjà triée donc les 1er)
        sorted_motifs = sorted_motifs[:top]
    
    # Reconstruire en dictionnaire
    motifsfreq = dict(sorted_motifs)

    return motifsfreq

    
test_motifs_top = getTopMotifs(test_motifs, 2, decroissant = True)
    
print(test_motifs_top)
```

```python
def searchGivenMotif(motifsDict, motifSpecifique, decroissant = True):
    """
    Cherche un motif specifique dans un dictionnaire de motifs trouvés
    entrée motifsDict : dictionnaire de motifs, clé = motif, valeur = fréquence d'observation
    entrée motifSpecifique: un motif specifique à chercher
    entrée decroissant : bool, si True, le dictionnaire est trié par ordre décroissant de valeur
    sortie fréquence : la fréquence du motif
    sortie ranking : dans quelle position le motif a été trouvé
    >>>searchGivenMotif(test_motifs, "TAT")
    (2, 2)
    """
    motifsList = sorted(motifsDict.items(), key = lambda x: x[1], reverse = decroissant)
    # motifsList est une liste de tuples (motif, freqc)

    resultat = next((t for t in motifsList if motifSpecifique in t[0]), None) # le tuple qui est de la forme (motifSpecifique,_)
    if resultat == None : # si le motif n'est pas dans le dictionnaire 
        print(motifSpecifique + "N'est pas présent")
        return 0, 0
        
    index = motifsList.index(resultat)
        
    ranking = index + 1 # le rang commence à 0
    frequence = motifsList[index][1]
    
    return ranking, frequence
 
searchGivenMotif(test_motifs, "TAT", decroissant = True)
```
